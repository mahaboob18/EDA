```python
# Set plot style
sns.set_theme(style="whitegrid")

# Define numerical columns to check for outliers
numerical_columns = ['realSum', 'guest_satisfaction_overall', 'cleanliness_rating', 'dist', 'metro_dist']

# Plot boxplots for these columns
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(15, 12))
fig.suptitle("Boxplots for Outlier Detection", fontsize=16)

for i, col in enumerate(numerical_columns):
    row, col_idx = divmod(i, 2)
    sns.boxplot(data=combined_data, x=col, ax=axes[row, col_idx])
    axes[row, col_idx].set_title(f"{col}")

# Hide the empty subplot if the number of plots is odd
if len(numerical_columns) % 2 != 0:
    axes[-1, -1].axis("off")

plt.tight_layout()
plt.show()
```

## The boxplots reveal potential outliers in several numerical columns:

1.realSum (price): Significant outliers, likely due to luxury accommodations or extreme pricing.

2.guest_satisfaction_overall: Mostly clustered at high ratings, with a few lower outliers.


3.cleanliness_rating: Majority are high, but a few low ratings appear as outliers.


4.dist (distance from city center): Some listings are located far from the city center.

5.metro_dist: A few properties are unusually far from metro stations.


```python
def remove_out(df,columns):
    for col in df[col]:
        q1=df[col].quantile(0,25)
        q3=df[col].quantiles(0.75)
        iqr=q3-q1
        lower_bound=q1-1.5*iqr
        upper_bound=q3+1.5*iqr
        df=df[(df[col]<=lower_bound)&(df[col]>=upper_bound)]
    return df  
        
```

# 5. Weekday vs. Weekend Analysis

 Price Fluctuation: Analyze the difference in average and median prices between
weekdays and weekends for each city.

 Room Availability: Explore how room capacity (person_capacity) changes across
day types.

 Satisfaction Variability: Investigate whether guest satisfaction scores differ between

```python
# Group data by city and day_type to calculate mean and median prices
price_fluctuation = combined_data.groupby(['city', 'day_type'])['realSum'].agg(['mean', 'median']).reset_index()

# Visualize average and median prices
plt.figure(figsize=(12, 6))
sns.barplot(data=price_fluctuation, x="city", y="mean", hue="day_type")
plt.title("Average Price: Weekdays vs. Weekends by City")
plt.xlabel("City")
plt.ylabel("Average Price (realSum)")
plt.xticks(rotation=45)
plt.legend(title="Day Type")
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(data=price_fluctuation, x="city", y="median", hue="day_type")
plt.title("Median Price: Weekdays vs. Weekends by City")
plt.xlabel("City")
plt.ylabel("Median Price (realSum)")
plt.xticks(rotation=45)
plt.legend(title="Day Type")
plt.show()
```

### For Guests
